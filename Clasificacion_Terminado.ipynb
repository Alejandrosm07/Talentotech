Acerca del conjunto de datos

Este conjunto de datos proporciona una recopilación completa de información sobre distintos teléfonos inteligentes, lo que permite un análisis detallado de sus especificaciones y precios. Abarca una amplia gama de teléfonos inteligentes, que abarcan diversas marcas, modelos y configuraciones, lo que lo convierte en un recurso valioso para investigadores, analistas de datos y entusiastas del aprendizaje automático interesados ​​en la industria de los teléfonos inteligentes.

Los datos proceden del sitio web español PC Componentes. Los datos se recopilaron mediante Power Automate.

Campos incluidos:

Nombre del teléfono inteligente : el identificador único o nombre del modelo del teléfono inteligente.

Marca: Marca Smartphone.

Modelo: Modelo de la marca Smartphone.

RAM (Memoria de acceso aleatorio): La cantidad de memoria disponible para realizar múltiples tareas.

Almacenamiento: capacidad del smartphone.

Color: Color del smartphone.

Gratis: Sí/No si el teléfono inteligente está vinculado a un contrato de una compañía celular.

Precio: el costo del teléfono inteligente en la moneda respectiva. Al utilizar este conjunto de datos, los investigadores y analistas pueden explorar patrones, tendencias y relaciones entre las especificaciones de los teléfonos inteligentes y su precio. Sirve como un excelente recurso para tareas como la predicción de precios, el análisis de mercado y la comparación de diferentes configuraciones de teléfonos inteligentes. Ya sea que esté interesado en identificar las opciones más rentables o en comprender el impacto de componentes de hardware específicos en los precios de los teléfonos inteligentes, este conjunto de datos ofrece abundantes posibilidades para una exploración en profundidad.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Seleccionar características y etiquetas
X = smartphones[['RAM', 'Storage', 'Price']]  # Características
y = smartphones['Brand']  # Etiqueta

# Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)

# Normalizar las características
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Crear y entrenar el modelo SVM
model = SVC()
model.fit(X_train, y_train)

# Predecir con el modelo
pred = model.predict(X_test)

# Evaluar el modelo
train_score = model.score(X_train, y_train)
print(f'Train score of trained model: {train_score * 100:.2f}%')

test_score = model.score(X_test, y_test)
print(f'Test score of trained model: {test_score * 100:.2f}%')

# Crear la matriz de confusión
conf_matrix = confusion_matrix(y_test, pred)
print(f'Confusion Matrix: \n{conf_matrix}\n')

# Visualizar la matriz de confusión
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()

```
