```python
import pandas as pd

# Load the iris dataset, specifying that there is no header row
iris = pd.read_csv("iris.data", header=None)

# Add column names to the DataFrame
iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

iris
```

```python
# Calculate the average of the measurements for each species
setosa_avg = iris[iris['species'] == 'Iris-setosa'].mean()
versicolor_avg = iris[iris['species'] == 'Iris-versicolor'].mean()
virginica_avg = iris[iris['species'] == 'Iris-virginica'].mean()
```

```python
# Define a function to calculate the Manhattan distance
def manhattan_distance(a, b):
  return sum(abs(val1-val2) for val1, val2 in zip(a,b))

# Define a function to calculate the Euclidean distance
def euclidean_distance(a, b):
  return sum((val1-val2)**2 for val1, val2 in zip(a,b))**0.5
```

```python
# Calculate the Manhattan and Euclidean distances between each pair of averages
manhattan_sv = manhattan_distance(setosa_avg, versicolor_avg)
manhattan_sv = manhattan_distance(setosa_avg, virginica_avg)
manhattan_vv = manhattan_distance(versicolor_avg, virginica_avg)

euclidean_sv = euclidean_distance(setosa_avg, versicolor_avg)
euclidean_sv = euclidean_distance(setosa_avg, virginica_avg)
euclidean_vv = euclidean_distance(versicolor_avg, virginica_avg)
```

```python
# Create a distance matrix
distances = pd.DataFrame({
    'Species Pair': ['Setosa-Versicolor', 'Setosa-Virginica', 'Versicolor-Virginica'],
    'Manhattan Distance': [manhattan_sv, manhattan_sv, manhattan_vv],
    'Euclidean Distance': [euclidean_sv, euclidean_sv, euclidean_vv]
})
```
